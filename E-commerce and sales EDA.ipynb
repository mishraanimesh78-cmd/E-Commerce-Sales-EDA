# Customer Segment

```python
segments=df['Customer_Segment'].value_counts().reset_index()
segments
```

```python
plt.figure(figsize=(4,3))
plt.bar(segments['Customer_Segment'],segments['count'],color='black',width=0.3)
plt.xlabel('Customer Segment',color='black')
plt.ylabel('Customer count',color='black')
plt.grid(color='red')
plt.show()
```

# Region Count

```python
Region=df['Region'].value_counts().reset_index()
Region
```

```python
c=np.arange(len(Region['Region']))
plt.figure(figsize=(4,3))
plt.xticks(c,Region['Region'],rotation=90)
plt.bar(Region['Region'],Region['count'],color='black',width=0.3)
plt.xlabel('Region',color='purple')
plt.ylabel('count',color='purple')
plt.grid(color='red')
plt.show()
```

```python
salesbyproduct1=df.groupby('Product_Category')[['Total_Sales']].sum().reset_index()
salesbyproduct1
```

```python
plt.figure(figsize=(5,4))
plt.bar(salesbyproduct1['Product_Category'],salesbyproduct1['Total_Sales'],color='black',width=0.5)
plt.xlabel('Product_Category',color='purple')
plt.ylabel('Total_Sales',color='purple')
plt.grid(color='red')
plt.plot(salesbyproduct1['Total_Sales'],color='blue',marker='.')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.pie(salesbyproduct1['Total_Sales'])
plt.legend(bbox_to_anchor=(1.05,1.0),labels=salesbyproduct1['Product_Category'])
plt.show()
```

# Region Wise Product and their Sales and Profit

```python
df=pd.read_csv("E:\\Python\\DATA SET\\global_ecommerce_sales.csv")
```

```python
Data1=df.where(df['Region']=='North America')
Data1.dropna(inplace=True)
```

```python
Productsales1=Data1.groupby('Product_Name').agg({'Total_Sales':'sum','Profit':'sum'}).reset_index()
Productsales1
```

```python
melted=Productsales1.melt(id_vars='Product_Name',value_vars=['Total_Sales','Profit'],
                          var_name='Metric',value_name='Value')
plt.figure(figsize=(10,6))
sns.barplot(data=melted,x="Product_Name",y='Value',hue='Metric')
plt.xticks(rotation=90)
plt.title("Total Sales and Profit by Product")
plt.show()
```
